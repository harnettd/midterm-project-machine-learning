```python
import pandas as pd
import json
import os
```

```python
df = pd.read_json('AK_Juneau_0.json')
data_dict = df.to_dict(orient='records')
data_dict
```

```python
all_data = None

def combined_json_to_df(df1, df2, df3):
    global all_data

    # checking if files exist
    if not os.path.exists(df1) or not os.path.exists(df2): 
        raise FileNotFoundError("One or both Files not Found")
    #loading data from json
    with open(df1, 'r') as d_file1: 
        data1 = json.load(d_file1)

    with open(df2, 'r') as d_file2:
        data2 = json.load(d_file2)
        
    with open(df3, 'r') as d_file3:
        data3 = json.load(d_file3)
        
    first_data = pd.DataFrame(data1)
    second_data = pd.DataFrame(data2)
    third_data = pd.DataFrame(data3)

    #combining two dataframes
    combined_data  = pd.concat([first_data, second_data, third_data], ignore_index = True)
    
    all_data = combined_data

df1 = 'WY_Cheyenne_0.json'
df2 = 'WY_Cheyenne_1.json'
df3 = 'WY_Cheyenne_2.json'

combined_json_to_df(df1, df2, df3)
```
