```python
import pandas as pd
```

```python
columns = ['property_id', 'status', 'list_date', 'year_built', 'sold_date', 'sold_price','is_price_reduced'
           , 'baths', 'lot_sqft', 'sqft', 'garage', 'stories', 'beds', 'type', 'state', 'fireplace', 'dishwasher', 
    'laundry_room', 'master_bedroom', 'swimming_pool', 'water_view', 'lake', 'investment_opportunity', 'dining_room', 'senior_community']

df = pd.read_csv('housing_data_0.csv', usecols = columns)
```

```python
df
```

```python
def clean_and_convert(df):
    # Clean the date column to return dd-mm-yy format
    df['list_date'] = pd.to_datetime(df['list_date'], errors='coerce').dt.strftime('%d-%m-%y')
    df['sold_date'] = pd.to_datetime(df['sold_date'], errors='coerce').dt.strftime('%d-%m-%y')

    # Convert 'year_built' and columns to integer data type
    df['sold_price'] = pd.to_numeric(df['sold_price'], errors='coerce').astype('Int64')
   
    # Calculate the average of non-null 'year_built' values
    avg_year_lot_sqft = df['lot_sqft'].dropna().mean()
    
    # Replace null values in 'year_built' column with the average
    df['year_built'].fillna(avg_year_lot_sqft, inplace=True)
    
    return df
```

```python
df = clean_and_convert(df)
```
